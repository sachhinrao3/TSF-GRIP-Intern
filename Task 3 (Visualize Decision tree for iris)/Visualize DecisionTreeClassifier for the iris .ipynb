```python
#Importing the necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams["figure.figsize"] = (12, 7)
import warnings
warnings.filterwarnings('ignore')

from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

## Import the dataset

```python
iris = pd.read_csv('C:/Users/admin/Documents/GRIP Internship/Task 3 (Visualize Decision tree for iris)/Iris.csv', index_col='Id')
iris.head(6)
```

```python
# Shape of the dataset
print('Shape of the dataset:',iris.shape)
```

```python
# Unique labels in target feature
iris.Species.unique()
```

```python
iris.describe()
```

## Convert categorical feature into numerical of target variable

```python
cleanup_nums = {"Species" : {'Iris-setosa':0,
                        'Iris-versicolor':1,
                        'Iris-virginica':2}}
iris_df = iris.replace(cleanup_nums)

iris_df.head()
```

```python
# Our target feature
iris_df.Species.unique()
```

# Split the Data

```python
X= iris_df.drop('Species', axis=1)
y = iris_df['Species']
```

```python
# Fit the classifier with default hyper-parameters
clf = DecisionTreeClassifier(random_state=51)
model = clf.fit(X, y)
```

# Print Text Representation

```python
# Install required libraries
# !pip install pydotplus
# !apt-get install graphviz -y
```

```python
text_representation = tree.export_text(clf)
print(text_representation)
```

```python
# # We can save the Chart by
# with open("decistion_tree.log", "w") as fout:
#     fout.write(text_representation)
```

# Visualize DecisioinTree with plot_tree

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=X.columns,  
                   class_names=iris.Species.unique(),
                   filled=True)
```

```python
## To save the image of decision tree in file
# fig.savefig("decistion_tree.png")
```

# Prediction Using Decision Tree

```python
# Split the data using train_test_split method
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=51)
```

```python
# Fit the model on training data
clf.fit(X_train, y_train)
```

```python
# Predict the test data
y_pred= clf.predict(X_test)
```

###  Compare the Actual and predicted data

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head(10)
```

### Check the Confusion Matrix and Accuracy Score

```python
# Import the libraries to check the accuracy of our model
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, plot_confusion_matrix
```

```python
cf= confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)
score= accuracy_score(y_test, y_pred)

print('Confusion matrix for our model: \n',cf)
print(cr)
print('======================================')
print('Accuracy score for our model is:',score*100)
```

```python
disp = plot_confusion_matrix(model, X_test, y_test,
                                 display_labels=X.columns,
                                 cmap=plt.cm.Blues,
                                )
# disp.ax_.set_title(title)
```

__After being fitted, the model can then be used to predict the class of samples.__

```python
pred= clf.predict([[2., 2., 2., 2.]]) # 'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2
print('Predicted iris type is:', pred )
print('where: Iris-setosa:0', 'Iris-versicolor:1', 'Iris-virginica:2')
```
